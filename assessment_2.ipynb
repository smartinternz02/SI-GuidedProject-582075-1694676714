```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
data=sns.load_dataset('car_crashes')
```

```python
sns.scatterplot(x="total",y="speeding",data=data)
```

inference : The scatter plot shows a weak positive correlation between speeding  and the total of speeding-related accidents.
This suggests that in states with increasing amount of speeding increases total accidents

```python
sns.scatterplot(x="alcohol",y="not_distracted",data=data)
```

 From the scatter plot, there doesn't appear to be a strong, obvious relationship between alcohol consumption and the number of not distracted drivers. The data points are scattered across the plot, suggesting that factors other than alcohol consumption may influence driver distraction rates.

```python
sns.lineplot(x="total",y="alcohol",data=data,ci=None)
```

There appears to be a positive relationship between the total number of car crashes and alcohol consumption. In general, as alcohol consumption increases, the total number of car crashes tends to increase as well.

```python
sns.distplot(data["no_previous"])
```

The distribution of "no_previous" crashes is right-skewed, indicating that the majority of states have a relatively low number of drivers with no previous accidents.

```python
sns.relplot(x="total",y="speeding",data=data,hue="abbrev")
```

 There appears to be a positive correlation between total crashes and speeding fatalities. States with higher total crash rates also tend to have higher speeding fatality rates.
 This suggests that speeding may contribute to a significant portion of total crashes.

```python
sns.jointplot(x="total",y="speeding",data=data)
```

The scatter plot in the center of the joint plot suggests a positive correlation between the total number of car crashes and the number of speeding-related crashes. States with a higher total number of crashes tend to have more crashes related to speeding.

```python
sns.histplot(data['total'], bins=20, kde=True)
plt.xlabel('Total Car Crashes')
plt.ylabel('Frequency')
```

The histogram shows the distribution of total crashes. Most states have a relatively low number of total crashes, with a peak around 15-20 crashes.

```python
sns.pairplot(data)
plt.show()
```

this shows all types of data visualization graphs with all parameters

```python
plt.figure(figsize=(10, 6))
sns.barplot(x='abbrev', y='total', data=data)
plt.xlabel('State Abbreviation')
plt.ylabel('Average  Car Crashes')
plt.title('Average Car Crashes by  the State')
plt.xticks(rotation=90)
plt.show()
```

The bar plot shows the total number of car crashes for each state represented by its abbreviation. It is evident that there is significant variability in car crash rates among different states. Some states have notably higher total car crashes than others.

```python
sns.boxplot(data=data, x="speeding", y="total")
plt.xlabel("Speeding Fatalities")
plt.ylabel("Total Crashes")
plt.title("Speeding Fatalities vs. Total Crashes")
plt.show()
```

The box plot demonstrates that states with a higher number of speeding fatalities tend to have a wider range of total crashes, with some outliers having significantly higher crash rates.

```python
state_count = data['abbrev'].value_counts()
```

```python
plt.figure(figsize=(10, 12))
plt.pie(state_count, labels=state_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of total States in the car_crashes Dataset')
plt.show()
```

The pie chart reveals that the distribution of states in the "car_crashes" dataset is not evenly balanced. Some states have a more significant representation in the dataset, while others are less common.
